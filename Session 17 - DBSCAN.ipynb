# DBSCAN

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels as sm
```

```python
from sklearn.datasets import make_blobs

X,_=make_blobs(n_samples=1000,n_features=2,
              centers=4,random_state=10)
X
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4, random_state=10)
kmeans.fit(X)

# plt.figure(figsize=(10,8))
plt.scatter(X[:,0],X[:,1],c=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:,0],
           kmeans.cluster_centers_[:,1],
           c='r', marker='*',s=200);
```

## Drawback of K means

* Only Spherical clustering

* Need to know K in advance

* Unduly influenced by outliers

## DBSCAN

Density Based Spatial Clustering of Applications with Noise



```python
from sklearn.cluster import DBSCAN

db=DBSCAN(eps=0.5, min_samples=5)
db.fit(X)
```

```python
plt.scatter(X[:,0],X[:,1],c=db.labels_);
```

```python
db=DBSCAN(eps=0.2, min_samples=5)
db.fit(X)
plt.scatter(X[:,0],X[:,1],c=db.labels_);
```

## Prediction using KNeighborsClassifier

```python
X_new= db.components_
y=db.labels_[db.core_sample_indices_]
```

```python
X_new
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test,y_train,y_test=train_test_split(X_new,y,test_size=0.2, random_state=10)

X_train.shape, X_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
kn=KNeighborsClassifier(n_neighbors=20)
kn.fit(X_train,y_train)
```

```python
from sklearn.metrics import classification_report

report=classification_report(y_test,kn.predict(X_test))
print(report)
```

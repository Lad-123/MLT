# Performance Metrics of Classification 

Let X1, X2, ...., Xn be n number of features.

Z = b0 + b1 X1 +....+ bn Xn

f(Z) = e^Z / (1+e^Z) 

f(Z)  lies in between 0 and 1 for every value of Z

Such a function is called the logistic funtion or Sigmoid fucntion.



f(Z) is the probability

If 0<f(Z)<0.5 then Z==> belongs to the class 0.

If 0.5< f(Z)< 1 then Z==> belongs to the class 1.

$f(Z) = \frac{e^{Z}} { 1+ e^{Z}}$ 

## Logistic Regression for German Credit problem

### Sourcing the data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Data set

gc=pd.read_csv('https://online.stat.psu.edu/onlinecourses/sites/stat508/files/german_credit.csv')
```

```python
gc
```

```python
gc.info()
```

## Preprocessing of the data

```python
# Target

y = gc['Creditability']
y
```

```python
y.unique()
```

```python
# Only two classes- binary classification.
```

```python
# Features

X=gc.drop(['Creditability'],axis=1)
X
```

```python
X.shape
```

```python
X.columns
```

```python
category_cols=['Account Balance',
       'Payment Status of Previous Credit', 'Purpose', 
       'Value Savings/Stocks', 'Length of current employment',
       'Instalment per cent', 'Sex & Marital Status', 'Guarantors',
       'Duration in Current address', 'Most valuable available asset',
        'Concurrent Credits', 'Type of apartment',
       'No of Credits at this Bank', 'Occupation', 'No of dependents',
       'Telephone', 'Foreign Worker']
```

```python
len(category_cols)
```

```python
# Converting categorical features to numeric

X_1 = pd.get_dummies(X,columns=category_cols)
```

```python
X_1.shape
```

```python
# Adding constant

import statsmodels.api as sm
X_1=sm.add_constant(X_1)
X_1.shape
```

```python
X_1.iloc[0:10,0:20]
```

```python
X_1.iloc[0:10,20:40]
```

```python
X_1.iloc[0:10,40:60]
```

```python
X_1.iloc[0:10,60:]
```

## Splitting to train and test

```python
from sklearn.model_selection import train_test_split

X_train_1,X_test_1,y_train_1,y_test_1=train_test_split(
     X_1,y,test_size=0.2,random_state=10)

X_train_1.shape,X_test_1.shape,y_train_1.shape,y_test_1.shape
```

## Building the model

```python
logR_1=sm.Logit(y_train_1,X_train_1)

# Fit the model
logR_1=logR_1.fit()
```

## Diagnosing the model

```python
logR_1.summary2()
```

```python

# Features with p<0.05


imp_features_1= ['Duration of Credit (month)','Credit Amount','Age (years)']
```

```python
# Update the feature set

X_2=X_1[imp_features_1]
```

```python
X_2
```

## Continue with revise feature set

```python
X_train_2,X_test_2,y_train_2,y_test_2=train_test_split(
     X_2,y,test_size=0.2,random_state=10)

X_train_2.shape,X_test_2.shape,y_train_2.shape,y_test_2.shape
```

```python
# Building the model
logR_2= sm.Logit(y_train_2,X_train_2)
logR_2=logR_2.fit()
logR_2.summary2()
```

```python
# Singnificant features

imp_features_2=['Duration of Credit (month)','Age (years)']
```

```python
X_3=X_2[imp_features_2]
```

```python
X_3
```

## Continue with revised feature set

```python
X_train_3,X_test_3,y_train_3,y_test_3=train_test_split(
     X_3,y,test_size=0.2,random_state=10)

X_train_3.shape,X_test_3.shape,y_train_3.shape,y_test_3.shape
```

```python
# Building the model
logR_3=sm.Logit(y_train_3,X_train_3)
logR_3=logR_3.fit()
logR_3.summary2()
```

```python
# Model building is complete

logR_3.params
```

###  The final LR model:

Z  =Duration of Credit (month) * -0.031934  + 

                Age (years) * 0.044396
    
    f(Z) = e^ Z / (1+ e^ Z)

## Prediction using the model

```python
y_pred=logR_3.predict(X_test_3)
y_pred
```

```python
y_test_3
```

```python
# A DF 

pred_df=pd.DataFrame({'Actual Class':y_test_3,
                      'Predicted Prob.':y_pred})
```

```python
pred_df
```

```python
# Adding a col to the DF

pred_df['Predicted Class']=pred_df['Predicted Prob.'].map(
    lambda x: 1 if x >0.5 else 0 )
```

```python
pred_df
```

## ROC Curve

Receiver Operating Characteristics Curve

Plotted by taking:

FPR (Sensitivity) ===> X- axis

TPR(1- Specificity) ===> Y -axis

```python
from sklearn.metrics import roc_curve

fpr,tpr,threshold=roc_curve(pred_df['Actual Class'],
                           pred_df['Predicted Prob.'])

plt.plot(fpr,tpr);
```

```python
plt.plot(fpr,tpr)
plt.plot([0,1],[0,1],c='r')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.show();
```

## ROC-AUC- Score



```python
from sklearn.metrics import roc_auc_score

score=roc_auc_score(pred_df['Actual Class'],
                   pred_df['Predicted Class'])

print(' The ROC-AUC-Score:',score)
```

## Finding the optimum classification cut-off

Many measures:

Youden's index: Max { Sensitivity + Specificity -1} 
 
              : Max { Sensitivity - (1- Specificity) }
              
              : Max { FPR - TPR }, actual diff
              
              
       

```python
# Create a new DF 

fpr_tpr= pd.DataFrame({'FPR': fpr,
                      'TPR': tpr,
                      'Threshold': threshold})

fpr_tpr
```

```python
# Creating the col of Diff

fpr_tpr['Diff']= np.abs(fpr_tpr['FPR']- fpr_tpr['TPR'])

fpr_tpr
```

```python
# Sorting

fpr_tpr.sort_values('Diff',ascending=False)
```

The threshold which provides the maximum DIff is : 0.526841	

```python
# Use the new cut-off 

pred_df['New Predicted Class']=pred_df['Predicted Prob.'].map(
    lambda x: 1 if x >0.526841 else 0 )
```

```python
pred_df
```

```python
# New Confusion matrix

new_cm=confusion_matrix(pred_df['Actual Class'],
                       pred_df['New Predicted Class'])

print(' The new CM:\n', new_cm)
```

```python
new_score=roc_auc_score(pred_df['Actual Class'],
                       pred_df['New Predicted Class'])
print(' The new score:', new_score)
```
